## 06. Creating Line Charts

```python
# Creating a subset
df = ords_prods_cust_all[:2000000]
```

#### Does this work? The answer is, well, no. Thatâ€™s because this subset wouldnâ€™t be representative of the entire population.
#### The answer is simpleâ€”sampling!

## 6.1 Sampling Data

```python
# Create a list holding True/False values to test np.random.rant() <=0.7, (70/30 split of the data)
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cust_all)) <= 0.7
```

```python
dev
```

```python
# Store 70% of the sample in the df big
big = ords_prods_cust_all[dev]
```

```python
# Store 30% of the sample in the df big
small = ords_prods_cust_all[~dev]
```

#### Note: The tilde you see in the second line negates the argument, in other words, it stores everything thatâ€™s not dev, or False

```python
# Check the achieved results
len(ords_prods_cust_all)
```

```python
len(big) + len (small)
```

#### Note: The exact same number of rows and we can assume that your subsetting procedure was at least successful in the technical sense.

```python
# Reducing the samples to only those columns necessary for the chart, namely â€œpricesâ€ and â€œorders_day_of_week.â€
df_2 = small[['orders_day_of_week','prices']]
```

```python
# Create Line Chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
plt.xlabel("Order day Of Week", fontsize=10)
plt.ylabel("Prices",fontsize=10)
plt.title("Development of the product's prices across the order days of the week",fontsize=12)
plt.tight_layout()  # Automatically adjust subplot parameters
```

```python
# Export the line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'orders_dow_prices.png'))
```
