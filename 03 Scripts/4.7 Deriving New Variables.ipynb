### 3.1.  Creating the “price_label” column in the 'orders_products_merged' dataframe

#### 3.1.a. Using 'If-Statements with User-Defined Functions' to create the “price_range” column for the subset of the first million rows in the 'orders_products_merged' dataframe

```python
# Create a subset of the first one million rows
df = df_ords_prods_merged[:1000000]
```

```python
df.head()
```

```python
# Define the function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return np.nan
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df
```

#### 3.1.b. Using 'If-Statements with the loc() Function' to create the “price_range” column  the entire 'orders_products_merged' dataframe

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# renaming the column 'price_range_loc' to 'price_range'
df_ords_prods_merged.rename(columns={'price_range_loc': 'price_range'}, inplace=True)
```

### 3.2.  Creating the “busiest day” column in the 'orders_products_merged' dataframe

```python
df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Using 'If-Statements with For-Loops'
result = []

for value in df_ords_prods_merged['orders_day_of_week']:
  if value == 0:
    result.append('Busiest day')
  elif value == 4:
    result.append('Least busy')
  else:
    result.append('Regularly busy')
```

### 3.3.  Rewriting  the “busiest day” column in the 'orders_products_merged' dataframe into “busiest days” (plural).

```python
result= []

for value in df_ords_prods_merged['orders_day_of_week']:
  if value == 0 or value == 1:
    result.append('Busiest days')
  elif value == 3 or value == 4:
    result.append('Slowes days')
  else:
    result.append('Regularly busy')
```

```python
result
```

```python
df_ords_prods_merged['busiest_day'] = result
```

```python
# renaming the column 'busiest_day' to 'busiest_days'
df_ords_prods_merged.rename(columns={'busiest_day': 'busiest_days'}, inplace=True)
```

### Direction: When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

### 3.4 Creating the “busiest_period_of_day” column in the 'orders_products_merged' dataframe

```python
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### Note: “Most orders”  =  9 AM-4 PM ;  “Fewest orders” = 11 PM - 6 AM  and  “Average orders” = the rest of the hours

```python
# Using 'If-Statements with For-Loops'
result_1 = []

for value in df_ords_prods_merged['order_hour_of_day']:
  if 9 <= value <= 16:
    result_1.append('Most orders')
  elif value == 23 or 0 <= value <= 6:
    result_1.append('Fewest orders')
  else:
    result_1.append('Average orders')
```

```python
result_1
```

```python
df_ords_prods_merged['busiest_period_of_day'] = result_1
```
